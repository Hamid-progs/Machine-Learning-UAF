## Introduction of Numpy Arrays

```python
l1 = [1,20,-300,4,5]
a = np.array(l1)

print("Datatype of Array: ",type(a))  # print type of array
print("Size: ",a.size)   # print size
print("Dimension: ",a.ndim)   # print dimension 
print("Shape: ",a.shape)  # print shape of array

print("np.array: ",a)

print("Sum: ",np.sum(a)) # print sum of array
print("Max: ",np.max(a)) # print max of array 
print("Min: ",np.min(a)) # print min of array 
print("Index of Max: ",np.argmax(a)) # print index of max 
print("Index of Min: ",np.argmin(a)) # print index of min
print("Mean: " , np.mean(a)) # print mean
```

```python
# append
a = np.append(a,230)
print(a)
```

### Inline Filtering without using loops

```python
# printing evern numbers
print(A[A%2 == 0])
```

```python
# print numbers which are even and also divisible on 10 
print(A[(A%2 == 0) & (A%10 == 0)])
```

```python
# print numbers which are even and divisible on 7 or 3
print(A[(A%2 == 0) & ((A%7 == 0) | (A%3 == 0))])
```
