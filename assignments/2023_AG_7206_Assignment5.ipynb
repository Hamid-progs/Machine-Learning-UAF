## <b>Q1: Must follow following steps:</b>
### Develop two 1-D integer arrays named A and B.

```python
print('Sum: ',A+B)
```

> #### <b>Find the difference of the two arrays.</b>

```python
print('Diffrence: ', A-B)
```

```python
print('Multiplication: ',A*B)
```

```python
print('Division: ',A/B)
```

```python
print('A^3: ',pow(A,3))
```

## <b>Q2: Must follow following steps:</b>

### Develop 2-D integer arrays named A.
### For example, if the last four digits are 5010, then the array size will be 50×10 OR if the last four digits are 5000, then the array size will be 50×50
### Let the last four digits of your AG number be 5010.
### Then the starting point is –5010 and the ending point is 5010.



```python
A = hamid.random.randint(-7206,7206,(72,6))
print(A)
```

### Perform the following operations:
> #### <b>Print Positive Even</b>

```python
print(A[(A > 0) & (A % 2 == 0)])
```

> #### <b>Print negative even</b>

```python
print(A[(A < 0) & (A % 2 == 0)])
```

> #### <b>Print Positive odd</b>

```python
print(A[(A > 0) & (A % 2 != 0)])
```

> #### <b>Print negative odd</b>

```python
print(A[(A < 0) & (A % 2 != 0)])
```

> #### <b>Print sum of all elements</b>

```python
print('Sum of Array "A": ',hamid.sum(A))
```

> #### <b>Print row wise sum</b>

```python
print('Row wise Sum: \n',hamid.sum(A,axis=1))
```

> #### <b>Print column wise sum</b>

```python
print('Column wise Sum: \n',hamid.sum(A,axis=0))
```

> #### <b>Print number may be even or dividable on 10</b>

```python
print(A[(A % 2 == 0) | (A % 10 == 0)])
```

> #### <b>Print number must be odd, positive and dividable 11</b>

```python
print(A[(A % 2 != 0) & (A > 0) & (A % 11 == 0) ])
```

> #### <b>Print total number of positive even</b>

```python
print('Size of Positive Even Numbers: ',A[(A > 0) & (A % 2 == 0)].size)
```

> #### <b>Print total number of negative even</b>

```python
print('Size of Negitive Even Numbers: ',A[(A < 0) & (A % 2 == 0)].size)
```

> #### <b>Print total number of even</b>

```python
print('Size of Even Numbers: ',A[A % 2 == 0].size)
```

> #### <b>Print total number of odd</b>

```python
print('Size of Odd Numbers: ',A[A % 2 != 0].size)
```
